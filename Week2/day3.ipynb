```python
import os
from dotenv import load_dotenv
import ollama
import gradio as gr
```

```python
system_message = "you are a helpful assistant "

```

```python
def chat(message, history, model="llama3.2"):
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    
    print("History is:")
    print(history)
    print("And messages is:")
    print(messages)

    # Ollama stream
    stream = ollama.chat(
        model=model,
        messages=messages,
        stream=True
    )

    response = ""
    for chunk in stream:
        response += chunk["message"]["content"] or ''
        yield response
```

```python
gr.ChatInterface( fn=chat, title="Ollama Chat with Llama3.2", description="Chat with Llama3.2 model using Ollama API", type="messages" ).launch() 
```

```python
def chat(message, history):
    """
    Ollama Llama3.2 stream chat fonksiyonu
    history: [{"role": "user/assistant", "content": "..."}]
    """
    # Mesajları hazırla
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    
    # Ollama stream
    stream = ollama.chat(model="llama3.2", messages=messages, stream=True)

    response = ""
    for chunk in stream:
        response += chunk["message"]["content"] or ''
        yield response
```

```python
gr.ChatInterface(
    fn=chat,
    title="Ollama Chat with Llama3.2",
    description="Chat with Llama3.2 model using Ollama API",
    type="messages"
).launch()
```
